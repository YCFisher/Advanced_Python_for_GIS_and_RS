```python
import pandas as pd
import arcgis
from arcgis.features import SpatialDataFrame 
from arcgis.gis import GIS
```

```python
gis = GIS("http://slustl.maps.arcgis.com/home", "cuaycongy_SLUSTL")
```

```python
equake_csv = pd.read_csv("all_week.CSV")
equake_csv.head()
```

```python
aug13_csv = pd.read_csv("all_week_aug_13_20.CSV")
aug13_csv.head()
```

```python
comb_df = equake_csv.append(ignore_index=True, other=aug13_csv)
```

```python
locations = comb_df.iloc[:,1:3].values.tolist()
```

```python
sdf = SpatialDataFrame(comb_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
```
