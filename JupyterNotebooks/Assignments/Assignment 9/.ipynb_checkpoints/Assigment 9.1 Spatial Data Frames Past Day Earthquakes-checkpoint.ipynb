```python
import pandas as pd
import arcgis
from arcgis.features import SpatialDataFrame 
from arcgis.gis import GIS
```

```python
gis = GIS("http://slustl.maps.arcgis.com/home", "cuaycongy_SLUSTL")
```

```python
equake_csv = pd.read_csv("all_day.CSV")
equake_csv.head()
```

```python
new_df = equake_csv.drop(columns=['nst', 'dmin', 'rms', 'net'])
new_df
```

```python
locations = new_df.iloc[:,1:3].values.tolist()
locations
```

```python
sdf = SpatialDataFrame(new_df,geometry=[arcgis.geometry.Geometry({'x':r[1], 'y':r[0], 
                    'spatialReference':{'wkid':4326}}) for r in locations])
sdf
```

```python
sdf1 = sdf[sdf['mag'] < 2.5]
sdf2 = sdf[(sdf['mag']>2.5) & (sdf['mag'] < 4)]
sdf3 = sdf[(sdf['mag']>4) & (sdf['mag'] < 5)]
sdf4 = sdf[sdf['mag'] > 5]
```

```python
sdf1
```

```python
sdf3
```

```python
sdf4
```

```python
sdf1.reset_index(inplace=True, drop=True)
sdf2.reset_index(inplace=True, drop=True)
sdf3.reset_index(inplace=True, drop=True)
sdf4.reset_index(inplace=True, drop=True)
```

```python
eq = gis.content.import_data(sdf4, title='Magnitude Greater than 5')
```

```python
eq_map = gis.map('South America',2)
eq_map
```

```python
eq_map.add_layer(eq, {"renderer":"Simple",
               "opacity": 0.75})
```

```python
webmap_properties = {'title':'Assignment 9.1 Spatial Data Frame Past Day Earthquakes Magnitude Greater than 5',
                     'snippet':'Webmap',
                     'tags':['SDF']}
eq_map.save(webmap_properties)
print("Done")
```
