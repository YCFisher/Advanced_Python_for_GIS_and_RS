```python
import requests
import pandas as pd

```

```python
year = '2015'
month = '1'
country = 'NOR'
```

```python
url = 'http://popdata.unhcr.org/api/stats/asylum_seekers_monthly.json?year=' + year + '&month=' + month + '&coa=' + country
data = requests.get(url)
as_json = data.json()
df = pd.DataFrame(as_json)
df
```

```python
most = df.sort_values(by=['value'], ascending=False)
most
```
